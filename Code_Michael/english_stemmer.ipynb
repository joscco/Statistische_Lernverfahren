```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
```

```python
import pandas as pd
import csv

csv_input = pd.read_csv("reviewEnglisch.csv", encoding="ISO-8859-1", sep = ';')
```

```python
import string
import re

def cleanText(text):
    translate_table = dict((ord(char), None) for char in string.punctuation)   
    text = text.translate(translate_table)
    text = re.sub('\d', '0', text)
    text = re.sub('0+', '0', text)
    return word_tokenize(text)
```

```python
ps.stem(csv_input["Englisch"][0])

listOfDocIds = []
listOfRawWords = []
listOfStemmedWords = []
docCounter = 1

for text in csv_input["Englisch"]:
    words = cleanText(text)
    
    for word in words:
        listOfDocIds.append(docCounter)
        listOfRawWords.append(word)
        listOfStemmedWords.append(ps.stem(word))
        
    docCounter = docCounter + 1
```

```python
listOfStemmedWords
```

```python
daten = pd.DataFrame(list(zip(listOfDocIds, listOfRawWords, listOfStemmedWords)), columns =['doc_id', 'cleaned_text', 'preprocessed_text']) 
```

```python
daten.to_csv("reviewEnglisch_preprocessed.csv", index=False, encoding="ISO-8859-1", quoting=csv.QUOTE_ALL)
```
